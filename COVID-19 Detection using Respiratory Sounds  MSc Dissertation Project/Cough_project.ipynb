# **Imputation of Missing Values for Categorical Fields with Mode and Conditional Logic**

Impute missing values in categorical columns using mode or specific conditions. Apply targeted imputations for fields like 'test_status' and 'test_date' based on logical rules

```python
#Mode Imputation for categorical fields
#Fill missing values in the columns with the most frequent value (mode)
df_cleaned['Locality'] = df_cleaned['Locality'].fillna(df_cleaned['Locality'].mode()[0])
df_cleaned['Returning_User'] = df_cleaned['Returning_User'].fillna(df_cleaned['Returning_User'].mode()[0])
df_cleaned['smoker'] = df_cleaned['smoker'].fillna(df_cleaned['smoker'].mode()[0])
df_cleaned['ctScan'] = df_cleaned['ctScan'].fillna(df_cleaned['ctScan'].mode()[0])
df_cleaned['Using_Mask'] = df_cleaned['Using_Mask'].fillna(df_cleaned['Using_Mask'].mode()[0])

#Conditional Imputation for 'test_status'
#Set 'test_status' to 'Not Taken Test' for rows where 'test_status' is missing and 'covid_status' indicates low risk or 'test_date' is missing
df_cleaned['test_status'] = df_cleaned.apply(lambda row: 'Not Taken Test' if pd.isna(row['test_status']) and
                             (row['covid_status'] in ['healthy', 'no_resp_illness_exposed', 'resp_illness_not_identified'] or pd.isna(row['test_date']))
                             else row['test_status'], axis=1)

#Impute 'testType' with a specific category where missing values indicate 'Not Taken'
df_cleaned['testType'] = df_cleaned['testType'].fillna('Not Taken')

#Conditional Imputation for 'test_date'
#Set 'test_date' to 'No Date' for rows where 'test_date' is missing and 'testType' is 'Not Taken'
df_cleaned.loc[(df_cleaned['test_date'].isna()) & (df_cleaned['testType'] == 'Not Taken'), 'test_date'] = 'No Date'

#Vaccination status: fill missing values as 'Not Vaccinated'
df_cleaned['Vaccination_status'] = df_cleaned['Vaccination_status'].fillna('Not Vaccinated')

#Confirm the changes by printing the count of missing values per column
print(df_cleaned.isna().sum())
```

# **Detecting Empty, Near-Silent, or Problematic Audio Files in Dataset**

Detect and list empty, near-silent, or problematic audio files in the specified directory, helping to ensure only valid cough audio samples are used for analysis. This function calculates the duration and checks the maximum amplitude of each audio file, flagging files that are either too short, nearly silent, or cause processing errors.

```python
import soundfile as sf

def check_empty_audio_files(root_dir, threshold=0.0001, min_duration=0.1):
    results = [] #Initialize an empty list to store results

    for folder_path, _, files in os.walk(root_dir):
        for file in files: #Check if the file has a valid audio format
            if file.endswith(('.wav', '.mp3', '.ogg', '.flac')):  #Add or remove audio formats as needed
                file_path = os.path.join(folder_path, file)
                try:
                    with sf.SoundFile(file_path) as audio_file:
                        duration = len(audio_file) / audio_file.samplerate #Calculate the duration of the audio file

                        if duration < min_duration: #Check if the audio file is too short
                            results.append({
                                'full_path': file_path,
                                'status': f'too short (duration: {duration:.2f}s)'
                            })
                        else:
                            #Read the entire file
                            audio_data = audio_file.read()

                            #Check if the max absolute amplitude is below the threshold
                            if np.max(np.abs(audio_data)) < threshold:
                                results.append({
                                    'full_path': file_path,
                                    'status': 'empty or near-silent'
                                })

                except Exception as e: #Handle any errors that occur while processing the audio file
                    results.append({
                        'full_path': file_path,
                        'status': f'error: {str(e)}'
                    })

    return pd.DataFrame(results)

#Specify the root directory to start searching from
root_directory = '/content/drive/MyDrive/cough_dataset/20200415'

#Run the check for empty or near-silent audio files
df = check_empty_audio_files(root_directory)

#Display results
if not df.empty:
    print("Empty, near-silent, or problematic audio files found:")
    pd.set_option('display.max_colwidth', None)  #Show full content of each column
    print(df)
    print(f"\nTotal files with issues: {len(df)}")
else:
    print("No empty or problematic audio files found.")
```

# **Audio Processing and Feature Extraction with Data Augmentation**
**Audio Augmentation Function**

This function introduces random audio augmentations to the waveform, with options for time stretching, pitch shifting, and adding noise. These augmentations simulate natural variations in cough sounds, helping the model generalize better for COVID classification

```python
import sys
import torch
import torchaudio
from tqdm.auto import tqdm
import concurrent.futures
import logging

def augment_audio(waveform, sample_rate):

    augmented = waveform

    #Time stretch (speed up/slow down) with a 50% probability
    if np.random.rand() < 0.5:
        stretch_factor = np.random.uniform(0.8, 1.2) #Random stretch factor between 0.8 and 1.2
        augmented = torchaudio.transforms.TimeStretch(stretch_factor)(augmented) #Apply time stretching

    #Pitch shift with a 50% probability
    if np.random.rand() < 0.5:
        n_steps = np.random.randint(-2, 3) #Randomly choose steps to shift pitch (-2 to +2)
        augmented = torchaudio.transforms.PitchShift(sample_rate, n_steps)(augmented) #Apply pitch shifting

    #Add random noise with a 50% probability
    if np.random.rand() < 0.5:
        noise = torch.randn_like(augmented) * 0.005 #Generate random noise
        augmented = augmented + noise #Add noise to the waveform

    return augmented
```

**Audio File Processing Function**

This function loads an audio file, applies optional augmentations, and extracts MFCC features, using padding or truncation to ensure a consistent length. The features, optimized for COVID cough detection, are prepared as 2D arrays suitable for model input.

```python
def process_audio_file(file_path, max_pad_length=100, n_mfcc=40, augment=False):
    try:
        logging.info(f"Processing file: {file_path}")
        waveform, sample_rate = torchaudio.load(file_path)

        if augment: #Apply augmentations if specified
            waveform = augment_audio(waveform, sample_rate)

        if torch.cuda.is_available(): #Move waveform to GPU if available
            waveform = waveform.cuda()

        # MFCC extraction
        n_fft = 4096 #Number of FFT points
        hop_length = n_fft // 4 #Hop length for frame shifting

        #Define the MFCC transformation
        mfcc_transform = torchaudio.transforms.MFCC(
            sample_rate=sample_rate,
            n_mfcc=n_mfcc,
            melkwargs={
                'n_fft': n_fft,
                'n_mels': 80, #Number of mel bands
                'hop_length': hop_length,
                'mel_scale': 'htk', #Mel scale type
            }
        )
        features = mfcc_transform(waveform) #Apply the MFCC transformation to the waveform

        features = features.cpu().numpy() #Convert to NumPy array for further processing

        #Padding or truncating the features to the desired length
        if features.shape[2] < max_pad_length:
            features = np.pad(features, ((0, 0), (0, 0), (0, max_pad_length - features.shape[2])), mode='constant') #Pad the features if they are shorter than max_pad_length
        else:
            features = features[:, :, :max_pad_length] #Truncate the features if they are longer than max_pad_length

        features_2d = features[0]  #From (1, n_mfcc, time) to (n_mfcc, time)

        logging.info(f"Successfully processed file: {file_path}")
        return features_2d

    except Exception as e:
        logging.error(f"Error processing {file_path}: {str(e)}")
        return None
```

**Load Audio Data Function**

This function loads and processes cough audio data in parallel, extracting MFCC features and associating each file with a COVID status label from metadata. Audio files are processed in batches with options for data augmentation, and labels are one-hot encoded to support multiclass classification.

```python
def load_audio_data(root_dir, metadata, max_pad_length=100, batch_size=500, n_mfcc=40, augment=False):
    """Load audio data and metadata using multiprocessing."""
    audio_files = []
    for root, _, files in os.walk(root_dir): #Traverse the directory tree to find all audio files
        for file in files:
            if file.endswith('.wav'): #Consider only .wav files
                audio_files.append(os.path.join(root, file))

    total_files = len(audio_files)
    logging.info(f"Total audio files found: {total_files}")
    print(f"Total audio files found: {total_files}")

    X = []
    y = []

    #Determine the number of workers to use for processing
    num_workers = max(1, os.cpu_count() // 2)
    logging.info(f"Using {num_workers} workers for processing")
    print(f"Using {num_workers} workers for processing")

    with concurrent.futures.ProcessPoolExecutor(max_workers=num_workers) as executor: #Process audio files in parallel
        future_to_file = {executor.submit(process_audio_file, file, max_pad_length, n_mfcc): file for file in audio_files} #Submit each audio file for processing and store the future results

        for future in tqdm(concurrent.futures.as_completed(future_to_file), total=total_files, desc="Processing audio files"): #Iterate over the completed futures
            file = future_to_file[future] #Get the file associated with the future
            try:
                result = future.result() #Retrieve the result of the processing
                if result is not None:
                    X.append(result)  #Now result is a 2D array (n_mfcc, time)

                    id_folder = os.path.basename(os.path.dirname(file)) #Extract the folder ID from the file path
                    covid_status = metadata.loc[metadata['id'] == id_folder]['covid_status'].values[0] #Retrieve the COVID status from the metadata based on the folder ID

                    #OneHot encode the target label
                    onehot_label = encoder.transform(np.array([[covid_status]]))[0]
                    y.append(onehot_label)

                    if len(X) >= batch_size: #Yield a batch of data if the batch size is reached
                        logging.info(f"Yielding batch of {len(X)} processed files")
                        yield np.array(X), np.array(y)  #X will be (batch_size, n_mfcc, time)
                        X = []
                        y = []
            except Exception as e:
                logging.error(f"Error processing {file}: {str(e)}")

    if X:
        logging.info(f"Yielding final batch of {len(X)} processed files")
        yield np.array(X), np.array(y)
```

**Logging and Data Processing Setup**

This script configures logging and uses multiprocessing to load and preprocess audio files in batches, with and without augmentation, for COVID detection. It tracks processing status, logs the shape of each batch. After processing all batches, it concatenates the data into final datasets one for original data and one for augmented data.

```python
#Set up logging to output to both file and console
logging.basicConfig(level=logging.INFO,
                    format='%(asctime)s - %(levelname)s - %(message)s',
                    handlers=[
                        logging.FileHandler("audio_processing.log"),
                        logging.StreamHandler(sys.stdout)
                    ])

#Configure tqdm to use the same output as logging
tqdm.write = logging.info

#Usage
#Specify the root directory to start searching from
root_directory = '/content/drive/MyDrive/cough_dataset/20200415'
batch_size = 500

logging.info("Starting audio data processing")
print("Starting audio data processing")

#Initialize lists to hold original data
X_original, y_original = [], []
for X_batch, y_batch in load_audio_data(root_directory, df_cleaned, batch_size=batch_size, n_mfcc=40, augment=False):
    X_original.append(X_batch)
    y_original.append(y_batch)
    logging.info(f"Processed batch: X shape {X_batch.shape}, y shape {y_batch.shape}")
    print(f"Processed batch: X shape {X_batch.shape}, y shape {y_batch.shape}")

#Combine original batches
X_original = np.concatenate(X_original)
y_original = np.concatenate(y_original)

#Initialize lists to hold augmented data
X_augmented, y_augmented = [], []
for X_batch, y_batch in load_audio_data(root_directory, df_cleaned, batch_size=batch_size, n_mfcc=40, augment=True):
    X_augmented.append(X_batch)
    y_augmented.append(y_batch)
    logging.info(f"Processed batch: X shape {X_batch.shape}, y shape {y_batch.shape}")
    print(f"Processed batch: X shape {X_batch.shape}, y shape {y_batch.shape}")

#Combine augmented data for training
X_augmented = np.concatenate(X_augmented)
y_augmented = np.concatenate(y_augmented)

logging.info("Finished processing all batches")
print("Finished processing all batches")


logging.info(f"Final X original shape: {X_original.shape}")  #X will be (n_samples, n_mfcc, max_pad_length)
logging.info(f"Final y original shape: {y_original.shape}")
print(f"Final X original shape: {X_original.shape}")
print(f"Final y original shape: {y_original.shape}")

logging.info(f"Final X augmented shape: {X_augmented.shape}")  #X will be (n_samples, n_mfcc, max_pad_length)
logging.info(f"Final y augmented shape: {y_augmented.shape}")
print(f"Final X augmented shape: {X_augmented.shape}")
print(f"Final y augmented shape: {y_augmented.shape}")

print("Script completed")  #Final debugging print statement
```

# **Preparing Data for Non-Linear Model**

This code flattens each audio sample's MFCC feature matrix to a 1D array, enabling compatibility with Random Forest Model and Support Vector Machine that require 2D input. It also converts one-hot encoded labels into single-column indices, necessary for models that don’t support multi-column label formats.

```python
#Reshape the audio data for compatibility with models expecting 2D input.
#Audio data is typically represented as 2D arrays (e.g., time and frequency) after feature extraction.
#reshape(-1) collapses the frequency/time dimension, flattening each audio sample into a 1D array.
X_train_nl = X_train.reshape(X_train.shape[0], -1)
X_test_nl = X_test.reshape(X_test.shape[0], -1)

#Convert one-hot encoded labels back to a single column of class indices.
#np.argmax finds the index of the maximum value in each row, converting the one-hot encoding to a single label.
#This is necessary because models like Random Forest require single-column labels rather than one-hot vectors.
y_train_nl = np.argmax(y_train, axis=1)
y_test_nl = np.argmax(y_test, axis=1)

print("Data reshaped for Non Linear Models.")
print(f"Training set shape: {X_train_nl.shape}")
print(f"Test set shape: {X_test_nl.shape}")
```

# **Reshaping Input Data and Preparing Labels for CNN Model Training**

Reshape the training and test datasets by adding a channel dimension, converting audio features into a format suitable for Convolutional Neural Networks (CNNs) as (samples, height, width, 1). Additionally, use np.argmax to convert one-hot encoded labels into integer labels, simplifying representation for model training. This prepares the data for effective CNN processing.

```python
#Reshape the data to add a channel dimension, which is necessary for CNNs
#Adding a channel dimension allows CNNs to treat audio features (e.g., spectrograms) as "images" with 1 channel (grayscale).
X_train = np.expand_dims(X_train, axis=-1)
X_test = np.expand_dims(X_test, axis=-1)

print(f"Reshaped training set: {X_train.shape}")
print(f"Reshaped test set: {X_test.shape}")

#Convert one-hot encoded labels to integer labels
#CNNs often require labels as integers rather than one-hot encoded vectors.
#np.argmax converts one-hot encoded labels (e.g., [0, 1, 0]) to single integers (e.g., 1).
y_train = np.argmax(y_train, axis=1)
y_test = np.argmax(y_test, axis=1)
```

# **Convolutional Neural Network**

This code defines and trains an improved Convolutional Neural Network (CNN) model for multi-class audio classification using Keras. The model employs several convolutional layers with batch normalization, dropout for regularization, and a softmax output layer for classification into eight categories. Training is enhanced with callbacks for learning rate adjustments and early stopping. After training, the model's performance is evaluated using accuracy metrics and a confusion matrix.

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, BatchNormalization, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt

#Automatically determine input shape and number of classes from training data.
def get_model_parameters(X_train, y_train):

    #Get input shape from training data
    input_shape = X_train.shape[1:]  #Exclude batch dimension

    #Get number of classes from labels
    if len(y_train.shape) == 2:  #If labels are one-hot encoded
        num_classes = y_train.shape[1]
    else:  #If labels are integer encoded
        num_classes = len(np.unique(y_train))

    return input_shape, num_classes

input_shape, num_classes = get_model_parameters(X_train, y_train)

# Build the improved CNN model for audio data
def build_improved_cnn_model(input_shape, num_classes):
    model = Sequential([
        Conv2D(32, kernel_size=(3, 3), activation='relu', padding='same', input_shape=input_shape), #First Convolutional Block
        BatchNormalization(),
        Conv2D(32, kernel_size=(3, 3), activation='relu', padding='same'),
        BatchNormalization(),
        MaxPooling2D(pool_size=(2, 2)),
        Dropout(0.25),

        Conv2D(64, kernel_size=(3, 3), activation='relu', padding='same'), #Second Convolutional Block
        BatchNormalization(),
        Conv2D(64, kernel_size=(3, 3), activation='relu', padding='same'),
        BatchNormalization(),
        MaxPooling2D(pool_size=(2, 2)),
        Dropout(0.25),

        Conv2D(128, kernel_size=(3, 3), activation='relu', padding='same'), #Third Convolutional Block
        BatchNormalization(),
        Conv2D(128, kernel_size=(3, 3), activation='relu', padding='same'),
        BatchNormalization(),
        MaxPooling2D(pool_size=(2, 2)),
        Dropout(0.25),

        Flatten(), #Fully Connected Layers
        Dense(256, activation='relu'),
        BatchNormalization(),
        Dropout(0.5),
        Dense(128, activation='relu'),
        BatchNormalization(),
        Dropout(0.5),
        Dense(num_classes, activation='softmax') #Output Layer with softmax activation for multi-class classification
    ])
    return model

#Build and compile the model
#Using Adam optimizer for adaptive learning rate
#sparse_categorical_crossentropy is used as the loss function for integer class labels
model = build_improved_cnn_model(input_shape, num_classes)
model.compile(optimizer=Adam(learning_rate=0.001),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

#Define callbacks for adaptive learning rate and early stopping
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=1e-6)
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

#Train the model
history = model.fit(X_train, y_train,
                    validation_data=(X_test, y_test),
                    batch_size=64, #batch_size: Number of samples per gradient update
                    epochs=50, #epochs: Number of times to iterate over the dataset
                    callbacks=[reduce_lr, early_stopping])

#Evaluate the model on the test set
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f"Test loss: {test_loss}")
print(f"Test accuracy: {test_accuracy}")

#Make predictions and generate a classification report
y_pred = model.predict(X_test)
y_pred_classes = y_pred.argmax(axis=1) #Convert prediction probabilities to class labels

print("Classification Report:")
print(classification_report(y_test, y_pred_classes))

print("Accuracy Score:")
print(accuracy_score(y_test, y_pred_classes))

#Plot training & validation accuracy values
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['accuracy', 'val_accuracy'], loc='upper left')

#Plot training & validation loss values
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['loss', 'val_loss'], loc='upper left')
plt.tight_layout()
plt.show()

#Confusion Matrix Heatmap
plt.figure(figsize=(8, 6))
cm = confusion_matrix(y_test, y_pred_classes)
sns.heatmap(cm, annot=True, cmap='Blues', fmt='g')
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

# **Transformer Model**

This code defines a Transformer model for audio classification, featuring a custom TransformerBlock with multi-head attention and feed-forward networks, as well as a PositionEmbedding layer. The model is configured with parameters like embedding dimensions and the number of heads, and it utilizes early stopping and learning rate reduction during training. After fitting the model on the transformed training data, it evaluates performance using accuracy metrics and classification reports.

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix

class TransformerBlock(layers.Layer):
    def __init__(self, embed_dim, num_heads, ff_dim, rate=0.1):
        super(TransformerBlock, self).__init__()
        self.att = layers.MultiHeadAttention(num_heads=num_heads, key_dim=embed_dim)
        self.ffn = keras.Sequential([
            layers.Dense(ff_dim, activation="relu"), #First dense layer with ReLU activation
            layers.Dense(embed_dim), #Second dense layer projecting back to embedding dimension
        ])
        self.layernorm1 = layers.LayerNormalization(epsilon=1e-6) #First layer normalization
        self.layernorm2 = layers.LayerNormalization(epsilon=1e-6) #Second layer normalization
        self.dropout1 = layers.Dropout(rate) #Dropout for attention output
        self.dropout2 = layers.Dropout(rate) #Dropout for feed-forward network

    def call(self, inputs, training=None):
        attn_output = self.att(inputs, inputs) #Self-attention
        attn_output = self.dropout1(attn_output, training=training)
        out1 = self.layernorm1(inputs + attn_output) #Add & normalize
        ffn_output = self.ffn(out1) #Feed-forward network
        ffn_output = self.dropout2(ffn_output, training=training)
        return self.layernorm2(out1 + ffn_output) #Add & normalize again
#Define a positional embedding layer
class PositionEmbedding(layers.Layer):
    def __init__(self, maxlen, embed_dim):
        super(PositionEmbedding, self).__init__()
        self.pos_emb = layers.Embedding(input_dim=maxlen, output_dim=embed_dim)

    def call(self, x):
        maxlen = tf.shape(x)[1] #Get the sequence length dynamically
        positions = tf.range(start=0, limit=maxlen, delta=1) #Generate position indices
        positions = self.pos_emb(positions) #Lookup positional embeddings
        return x + positions #Add positional embeddings to input
#Define the Transformer-based model
def build_transformer_model(maxlen, embed_dim, num_heads, ff_dim, num_transformer_blocks, mlp_units, dropout=0.1, mlp_dropout=0.1, num_classes=8):
    inputs = keras.Input(shape=(maxlen, embed_dim)) #Input tensor
    x = layers.LayerNormalization(epsilon=1e-6)(inputs) #Normalize input
    x = PositionEmbedding(maxlen, embed_dim)(x) #Add positional embedding
    initial_representation = x #Save initial embedding for residual connection

    for _ in range(num_transformer_blocks):#Add Transformer blocks
        x = TransformerBlock(embed_dim, num_heads, ff_dim, dropout)(x)

    x = layers.Add()([x, initial_representation]) #Add residual connection
    x = layers.GlobalAveragePooling1D()(x) #Global pooling to aggregate sequence information
    for dim in mlp_units:#Add MLP head for classification
        x = layers.Dense(dim, activation="relu")(x)
        x = layers.Dropout(mlp_dropout)(x)
    outputs = layers.Dense(num_classes, activation="softmax")(x) #Output layer with softmax for classification
    return keras.Model(inputs, outputs)

#Model configuration parameters
maxlen = 100 #Maximum sequence length
embed_dim = 40 #Embedding dimension
num_heads = 4 #Number of attention heads
ff_dim = 64 #Hidden layer size in feed-forward network
num_transformer_blocks = 6 #Number of Transformer blocks
mlp_units = [128, 64] #Dense layers in MLP head
mlp_dropout = 0.1 #Dropout rate in MLP head
num_classes = 8 #Number of output classes

model = build_transformer_model(
    maxlen=maxlen,
    embed_dim=embed_dim,
    num_heads=num_heads,
    ff_dim=ff_dim,
    num_transformer_blocks=num_transformer_blocks,
    mlp_units=mlp_units,
    dropout=0.1,
    mlp_dropout=mlp_dropout,
    num_classes=num_classes
)
#Compile the model with Adam optimizer and sparse categorical crossentropy loss
model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.0005),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"],
)
#Define learning rate scheduler and early stopping
reduce_lr = keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=1e-6)
early_stopping = keras.callbacks.EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

#Train the model
history = model.fit(
    X_train_transformer,
    y_train,
    batch_size=16,
    epochs=50,
    validation_data=(X_test_transformer, y_test),
    callbacks=[reduce_lr, early_stopping]
)

#Evaluate the model on test data
test_loss, test_accuracy = model.evaluate(X_test_transformer, y_test)
print(f"Test loss: {test_loss}")
print(f"Test accuracy: {test_accuracy:.4f}")

#Make predictions
y_pred = model.predict(X_test_transformer)
y_pred_classes = y_pred.argmax(axis=1)

#Print classification report for model performance
print(classification_report(y_test, y_pred_classes))

# Plot training history
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(['accuracy', 'val_accuracy'], loc='upper left')

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(['loss', 'val_loss'], loc='upper left')
plt.tight_layout()
plt.show()

#Confusion Matrix Heatmap
plt.figure(figsize=(8, 6))
cm = confusion_matrix(y_test, y_pred_classes)
sns.heatmap(cm, annot=True, cmap='Blues', fmt='g')
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```
